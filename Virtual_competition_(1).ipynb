HISTOGRAM

````python
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(10, 6))
sns.histplot(df['Glucose'], bins=30, kde=True)
plt.title('Histogram of Glucose Levels')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
````

The histogram shows a distribution of glucose levels that is slightly right-skewed, with a peak around 100. This suggests that most people have glucose levels around 100, but there are a few people with significantly higher levels. The overall spread of the data is moderate, indicating some variability in glucose levels. The right-skewness suggests that a portion of the population may be at risk for diabetes or other health complications associated with high glucose levels.

````python
plt.figure(figsize=(10, 6))
sns.histplot(df['BloodPressure'], bins=10, kde=True)
plt.title('Histogram of BloodPressure Levels')
plt.xlabel('BloodPressure Levels')
plt.ylabel('Frequency')
plt.show()
````

The histogram shows a normal distribution of blood pressure levels, with most people having levels around 60-70 mmHg. There are a few people with very low or very high blood pressure, which may indicate potential health risks. Overall, the distribution suggests a generally healthy population, but further analysis is needed to understand the specific risks and develop targeted interventions.

````python
plt.figure(figsize=(10, 6))
sns.histplot(df['SkinThickness'], bins=30, kde=True)
plt.title('Histogram of SkinThickness')
plt.xlabel('SkinThickness')
plt.ylabel('Frequency')
plt.show()

````

The histogram shows a right-skewed distribution of skin thickness, with most people having thinner skin and a few having significantly thicker skin. This distribution could suggest potential health implications or variations in population demographics



````python
plt.figure(figsize=(10, 6))
sns.histplot(df['DiabetesPedigreeFunction'], bins=30, kde=True)
plt.title('Histogram of DiabetesPedigreeFunction')
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Outcome')
plt.show()
````

The histogram shows a right-skewed distribution of diabetes pedigree function, with most people having lower values and a few having significantly higher values. This suggests that a small portion of the population may be at a higher risk of developing diabetes due to their family history.

BOX PLOT

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['BMI'])
plt.title('Box Plot of BMI')
plt.xlabel('BMI')
plt.show()

````

The box plot shows a right-skewed distribution of BMI, with most people having a BMI around 30. There are a few people with significantly higher BMIs, which may indicate potential health risks.

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['BloodPressure'])
plt.title('Box Plot of BloodPressure')
plt.xlabel('BloodPressure')
plt.show()
````

The box plot shows a distribution of blood pressure with a median around 70 mmHg and a relatively small spread. There are a few outliers on the higher end, indicating that a portion of the population may be at risk for hypertension. Overall, the distribution suggests a generally healthy population, but further analysis is needed to understand the specific risks and develop targeted interventions.

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Pregnancies'])
plt.title('Box Plot of Pregnancies')
plt.xlabel('Pregnancies')
plt.show()
````

````python
##SCATTER PLOT##



plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Glucose'], y=df['BMI'])
plt.title('Scatter Plot of Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
````

The scatter plot shows a weak positive correlation between glucose levels and BMI, suggesting that higher BMI is generally associated with higher glucose levels. However, the relationship is not perfectly linear, and other factors may also influence glucose levels.

````python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Glucose'], y=df['Insulin'])
plt.title('Scatter Plot of Glucose vs. Insulin')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
````

The scatter plot shows a strong positive correlation between glucose levels and insulin levels, suggesting that higher glucose levels are generally associated with higher insulin levels. However, there are individual variations and outliers, indicating that other factors may also influence this relationship.

````python
## CORRELATION MATRIX##


plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
````

The correlation matrix shows that glucose, insulin, BMI, and pregnancies are strongly associated with diabetes risk. However, there is also multicollinearity among some of these variables, which might need to be considered when building predictive models.

````python
## PAIR PLOT##


sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Insulin', 'Outcome']], hue='Outcome')
plt.show()
````

The pair plot provides a comprehensive overview of the relationships between the variables in the dataset. It reveals strong correlations between glucose, insulin, BMI, and the outcome of diabetes, suggesting that these factors are important predictors of disease risk. Additionally, the pair plot shows other relationships between the variables that may be relevant for understanding the underlying mechanisms of diabetes.

````python
#Multivariate Analysis#



## 3D PLOT##

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Glucose'], df['BMI'], df['Age'], c=df['Outcome'], cmap='coolwarm')
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
plt.title('3D Scatter Plot of Glucose, BMI, and Age')
plt.show()
````

The 3D scatter plot shows the relationships between glucose, BMI, and age. While the plot suggests potential clustering and relationships, further analysis is needed to draw definitive conclusions. Additional variables and statistical methods could provide a more comprehensive understanding of the factors influencing diabetes risk.
